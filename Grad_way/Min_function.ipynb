```python
import numpy as np
from json import loads, dumps
from copy import deepcopy
import sys
import os
from matplotlib.pyplot import plot 
from math import sin, cos
```

```python
class Minimize_Function:
    """
    THis class looking for minimize values of function
    As arguments class accept function in str format
    start_point, h, gamma have default values, but
    you can change this
    If you using mathematical operations in the led function,
    for example math.log(x) , math.cos(x), you must enter
    name of arguments in list format(['x', 'y_1', 'y_2'])
    and you can input start argument

    Example function:
    x**2+y*x+3*math.cos(z)+math.log(j)
    """

    def __init__(self, strfunc, start_point=3, h=0.134, gamma=0.012, name_arg=None, arg_val=None):
        self.func = strfunc
        # for comfort set translate in list
        self.arg = name_arg or list(self._finde_arg())
        if arg_val is not None:
            self.arg_values = np.array(arg_val)
        else:
            self.arg_values = np.array([start_point for i in range(len(self.arg))])
        self.h = h
        self.gamma = gamma

    @staticmethod
    def _load_funk_into_file():
        """
        This method for loading
        result from the file
        """
        if not os.path.exists("work.data"):
            return {}
        with open("work.data", 'r') as f:
            dic = f.read()
            if dic:
                return loads(dic)
            return {}

    def _finde_arg(self):
        """
        This function looking for
        arguments in self.strfunc
        and return set of arguments

        """
        rez = set()
        for i in self.func:
            if i.isalpha():
                rez.add(i)
        return rez

    def _save_funk(self, rez):
        """
        This function save result
        in file
        """
        result = {}
        work_info = self._load_funk_into_file()
        if work_info:
            work_info[self.func] = [rez, list(self.arg_values)]
            with open("work.data", "w") as f:
                f.write(dumps(work_info))
                f.close()
        else:
            
            result[self.func] = [rez, list(self.arg_values)]
            with open("work.data", 'w') as f:
                f.write(dumps(result))
                f.close()

    def _func_with_values(self):
        """
        This function replaces
        letters arguments on number arguments
        """
        prom = deepcopy(self.func)
        for i in range(len(self.arg)):
            prom = prom.replace(self.arg[i], str(self.arg_values[i]))
        return prom

    def gradient(self):

        """
        This function calculate gradient
        for each arguments
        """
        grad = []
        # This lab function with values
        func_val = deepcopy(self.func)
        """
        In this loop, the derivative
         is calculated for each argument
        """
        for i in range(len(self.arg)):
            working_str = deepcopy(self.func)
            """
            Instead of the argument by which the derivative
            is taken, the value is substituted str(self.arg_values[i] + self.h)
            Other arguments get values from the self.arg_values
            """
            working_str = working_str.replace(self.arg[i], str(self.arg_values[i] + self.h))
            func_val = func_val.replace(self.arg[i], str(self.arg_values[i]))
            for j in range(len(self.arg_values)):
                # skip the differentiated argument
                if self.arg[j] == self.arg[i]:
                    continue
                working_str = working_str.replace(self.arg[j], str(self.arg_values[j]))
            # make gradient vector
            try:
                grad.append(((eval(working_str) - eval(self._func_with_values())) / self.h))
            except NameError:
                print("Function not found, please import this")
                sys.exit()
        return np.array(grad)

    def get_func_min(self):
        num = 0
        # if file has result, will display values without calculation
        if self.func in self._load_funk_into_file():
            print(f"minimum function {self.func} = {self._load_funk_into_file()[self.func][0]}")
            return 
        while True:
            num += 1
            func_1 = eval(self._func_with_values())
            self.arg_values = self.arg_values - self.gamma * self.gradient()
            """
            Some function not defined with negative value 
            this except processes this
            """
            try:
                func_2 = eval(self._func_with_values())
            except Exception:
                # save result in file
                self._save_funk(str(func_1))
                return f"minimum function {self.func} = {func_1}\n" \
                       f" values of arguments {self.arg_values}"

            print(f" arguments values {self.arg_values}")

            print(f" number iteration {num}")
            if func_2 >= func_1:
                self._save_funk(func_1)
                break

        return f"minimum function {self.func} = {eval(self._func_with_values())}\n" \
               f" values of arguments {self.arg_values}"
```
